```python
import pandas as pd

top_ten_df = pd.read_csv("2019_CSN_State_Top_Ten_Report_Categories.csv", skiprows=2, skipfooter=4, engine='python')
top_ten_df.head()
```

```python
fraud_reports_df = pd.read_csv("2019_CSN_State_Fraud_Reports_and_Losses.csv", skiprows=2, skipfooter=4, engine='python')
fraud_reports_df.head()
```

```python
renamed_top_ten = top_ten_df.rename(columns={"# of Reports": "Reports"})
renamed_top_ten.head()
```

```python
renamed_fraud_reports = fraud_reports_df.rename(columns={"# of Reports": "ReportNumber", "% Reporting $ Loss": "ReportingLoss", "Total $ Loss": "TotalLoss", "Median $ Loss": "MedianLoss"})
renamed_fraud_reports.head()
```

```python
renamed_top_ten['Reports'] = renamed_top_ten['Reports'].str.replace(",", "").astype(float)
```

```python
renamed_top_ten['Percentage'] = renamed_top_ten['Percentage'].str.replace("%", "").astype(float)
renamed_top_ten.head()
```

```python
renamed_fraud_reports['ReportNumber'] = renamed_fraud_reports['ReportNumber'].str.replace(",", "").astype(float)
```

```python
renamed_fraud_reports['ReportingLoss'] = renamed_fraud_reports['ReportingLoss'].str.replace("%", "").astype(float)
```

```python
renamed_fraud_reports['TotalLoss'] = renamed_fraud_reports['TotalLoss'].str.replace(",", "")
```

```python
renamed_fraud_reports['TotalLoss'] = renamed_fraud_reports['TotalLoss'].str.replace("$", "")
```

```python
renamed_fraud_reports['MedianLoss'] = renamed_fraud_reports['MedianLoss'].str.replace("$", "").astype(float)
renamed_fraud_reports.head()
```

```python
renamed_top_ten.to_csv('top_ten.csv', index=False)
renamed_fraud_reports.to_csv('fraud_reports.csv', index=False)
```
